```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
sr = 44100.
lengthMs = 100.
samps = sr*lengthMs/1000.
n = np.linspace(0, samps, samps)
# t index array in seconds 
t = np.linspace(0, lengthMs/1000., samps)
```

```python
def aToDb(array):
    import math
    db=[]
    for i in range(len(array)):
        dbi = 20*(math.log10(np.clip(array[i],0.0001,1)))
        db.append(dbi)
    return db
```

```python
fig = plt.figure(figsize = (20, 10))
ax1= fig.add_subplot(111)
ax2= fig.add_subplot(122)
ax1.axis([0,0.11, 0, 1.1])
ax2.axis([0,0.11, -60, 10])
def efunct(n, d):
    sig = []
    for i in n:
        sig.append(math.exp(i*d))
    return sig

sig = efunct(t, -120.)
ax1.plot(t, sig)
sig = aToDb(sig)
ax2.plot(t, sig)
ax1.grid(True)
ax2.grid(True)
#print t
```
