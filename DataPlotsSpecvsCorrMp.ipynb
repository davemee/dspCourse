A function for parsing typical coll/dict max data to an array:

```python
def parseFrame(jsonFile):
    import json
    json_data=open(jsonFile)
    data = json.load(json_data)
    json_data.close()
    array1 = []
    for i in range (len(data)):
        array1.append(data[str(i)][0])
    return array1
```

```python
Spec = parseFrame("nSpec.json")
```

```python
Corr = parseFrame("nCorrel.json")
```

```python
CorrMp = parseFrame("nCorrel2.json")
```

```python
#Plotting the FFT data

fig = plt.figure()
ax = fig.add_subplot(111)

n = np.linspace(0, 22050, len(Spec))
plt.plot(n, Spec, "k-.", label="Spectrum")
plt.plot(n, Corr, "k--", label="AutoCorrelation")
plt.plot(n, CorrMp, "k", label="AutoCorrelation*2")
plt.title('One Frame');
plt.axis([500, 1500, 0, 1.2])
ax.set_xlabel('frequency')
ax.set_ylabel('Linear Ampitude')
plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
#plt.savefig('specVsCorr.png')
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True)
ax1.plot(n, Spec, "k", label = "spectrum")
ax1.set_title('Multiple autocorrelation passes')
ax2.plot(n, Corr, "k", label = "autocorrelation")
ax3.plot(n, CorrMp, "k", label = "autocorrelation*2")
# Fine-tune figure; make subplots close to each other and hide x ticks for
# all but bottom plot.
plt.axis([500, 1500, 0, 1.2])
f.subplots_adjust(hspace=0.2)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
ax.xaxis.set_ticks([1.,2.,3.,10.])
ax1.grid(True)
ax2.grid(True)
ax3.grid(True)
ax1.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
ax2.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
ax3.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
ax3.set_xlabel('frequency')
ax2.set_ylabel('Linear Ampitude')
plt.savefig('multiPass.png')
```
