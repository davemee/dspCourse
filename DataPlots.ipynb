A function for parsing typical coll/dict max data to an array:

```python
def parseFrame(jsonFile):
    import json
    json_data=open(jsonFile)
    data = json.load(json_data)
    json_data.close()
    array1 = []
    for i in range (len(data)):
        array1.append(data[str(i)][0])
    return array1
```

```python
noiseAndSineSpec = parseFrame("noiseAndMinus20dBsine1k.json")
```

```python
noiseAndSineCorr = parseFrame("noiseAndMinus20dBsine1kC.json")
```

```python
#Plotting the FFT data

fig = plt.figure()
ax = fig.add_subplot(111)

n = np.linspace(0, 22050, len(noiseAndSineSpec))
plt.plot(n, noiseAndSineSpec, "k--", label="Spectrum")
plt.plot(n, noiseAndSineCorr, "k", label="AutoCorrelation")
plt.title('One Frame');
plt.axis([500, 1500, 0, 1.2])
ax.set_xlabel('frequency')
ax.set_ylabel('Linear Ampitude')
plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
plt.savefig('specVsCorr.png')
```

```python
spectrum = parseFrame("Kuebel-spectrumFrame.json")
accAuto = parseFrame("Kuebel_Acc_autoCorrel.json")
```

```python
def aToDb(array):
    import math
    db=[]
    for i in range(len(array)):
        dbi = 20*(math.log10(np.clip(array[i],0.0001,1)))
        db.append(dbi)
    return db
```

```python
specdb=aToDb(spectrum)
accDb=aToDb(accAuto)
```

```python
#Plotting the FFT data

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(211, title='One Frame of the Spectrum')
ax2 = fig.add_subplot(212, title='Accumulated Spectrum of Autocorrelation')
n = np.linspace(0, 22050, len(spectrum))
ax.plot(n, specdb, "k", label="spectrum")
ax2.plot(n, accDb, "k", label="acc. Autocorrelation")
#ax.title("One Frame of the Spectrum");
#ax2.title('Accumulated Spectrum of Autocorrelation');
ax.axis([20, 22050, -80., 10.])
ax2.axis([20, 22050, -80., 10.])
ax.set_xlabel('frequency[Hz]')
ax2.set_xlabel('frequency[Hz]')
ax.set_ylabel('Ampitude[dB]')
ax2.set_ylabel('Ampitude[dB]')
ax.set_xscale('log')
ax2.set_xscale('log')
#ax.set_yscale('log')
#plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)

ax.grid(True)
ax2.grid(True)
plt.savefig('compare.png')
```
